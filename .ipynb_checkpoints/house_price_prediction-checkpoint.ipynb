# House Price Prediction Project

This project uses linear regression to predict house prices based on features such as size and location.  
It includes data preprocessing, feature scaling, model training, and evaluation.  
The results show a strong correlation between predicted and actual prices.

### Import Libraries:

```python
import pandas as pd
import numpy as np
from numpy import sqrt 
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
```

### Read Data:

```python
data = "housing.csv"
df = pd.read_csv(data)
```

### Data preparation:

```python
X = df[['Avg. Area Income', 'Avg. Area House Age',
        'Avg. Area Number of Rooms', 'Area Population']]

y = df['Price']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=45)
```

### Data Scaling (Standardization): 

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.fit_transform(X_test)
```

### Model building and training:

```python
model = LinearRegression()

model.fit(X_train, y_train)
```

### Evaluation:

```python
pred = model.predict(X_test)
```

```python
print('Mean absolute error : ', mean_absolute_error(y_test, pred))
print('Mean squared error : ', mean_squared_error(y_test, pred))
print('root Mean squared error : ', sqrt(mean_squared_error(y_test, pred)))
print('R2 : ', r2_score(y_test, pred))
```

### Show Results:

```python
plt.figure(figsize=(8, 6))
plt.scatter(y_test, pred, alpha=0.6)
plt.title('Actual vs Predicted House Prices')
plt.xlabel('Actual Prices')
plt.ylabel('Predicted Prices')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--')
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.plot(y_test.values[:50], label='Actual')
plt.plot(pred[:50], label='Predicted')
plt.legend()
plt.title("Actual vs Predicted Prices (first 50 samples)")
plt.show()
```

### Conclusion
The model shows strong predictive accuracy as the predicted values closely follow the actual house prices , with R2 = 0.92 indicating excellent performance.
